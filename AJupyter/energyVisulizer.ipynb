```python
import matplotlib.pyplot as plt

#open file
file = open('energy.txt','r')

#read from the file
text = file.readlines()


dataPoints = []

for line in text:
    dataPoints.append(float(line[:-1]))
    
time = list(range(len(dataPoints)))
#visulize
plt.plot(time,dataPoints)
plt.xlabel("Time in steps")
plt.ylabel("Energy")
plt.show()
```
