```python
import matplotlib.pyplot as plt
import numpy as np

#open file
file = open('energy.txt','r')

#read from the file
text = file.readlines()
dataPoints = []
for line in text:
    dataPoints.append(float(line[:-1]))
    
time = list(range(len(dataPoints)))
difference = np.diff(dataPoints)

#visulize
plt.rcParams['figure.figsize'] = [16,9]
plt.subplot(1,2,1)
plt.plot(time,dataPoints)
plt.xlabel("Time in steps")
plt.ylabel("Energy")
plt.title("Energy")

plt.subplot(1,2,2)
plt.plot(time[:-1],difference)
plt.xlabel("Time in steps")
plt.ylabel("Rate of Change")
plt.title("Diff. of Energy")
plt.savefig('plot.png')
plt.show()
```
