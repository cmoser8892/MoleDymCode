```python
import matplotlib.pyplot as plt
import numpy as np
import sys

#open file
file = open('../AData/potentialEnergy.txt','r')

#read from the file
text = file.readlines()
potentialEnergy = []
for line in text:
    potentialEnergy.append(float(line[:-1]))
    
file.close()
#
#open file
file = open('../AData/kineticEnergy.txt','r')
#read from the file
text = file.readlines()
kineticEnergy = []
for line in text:
    kineticEnergy.append(float(line[:-1]))

file.close()
#create Time
timestep = 1e-15
if(len(potentialEnergy) != len(kineticEnergy)):
    print("lenghts dont match")
    sys.exit()
else:
    lenght = len(potentialEnergy)
    time = np.arange(start = 0, stop = timestep*lenght, step = timestep)

#
plt.rcParams['figure.figsize'] = [16,9]    
#plt.plot(time,potentialEnergy)
plt.plot(time,kineticEnergy)
plt.title("kinetic energy to time")
plt.xlabel("time in fs")
plt.ylabel("energy in eV")
plt.show()
```
