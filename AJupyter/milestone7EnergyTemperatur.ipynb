```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sys
from scipy.signal import savgol_filter

#open file
file = open('../AData/potentialEnergy.txt','r')

#read from the file
text = file.readlines()
potentialEnergy = []
for line in text:
    potentialEnergy.append(float(line[:-1]))
    
file.close()
#
#open file
file = open('../AData/kineticEnergy.txt','r')
#read from the file
text = file.readlines()
kineticEnergy = []
for line in text:
    kineticEnergy.append(float(line[:-1]))

file.close()
#create Time
timestep = 1
if(len(potentialEnergy) != len(kineticEnergy)):
    print("lenghts dont match")
    sys.exit()
else:
    lenght = len(potentialEnergy)
    time = np.arange(start = 0, stop = timestep*lenght, step = timestep)

#filter data couse this looks way to messy
df = pd.DataFrame(dict(x=kineticEnergy))
kineticEnergyFiltered = df[["x"]].apply(savgol_filter, window_length=701, polyorder=2)

#plot
plt.ion()
plt.rcParams['figure.figsize'] = [16,9]    
#plt.plot(time,potentialEnergy)
plt.plot(time,kineticEnergy)
plt.plot(kineticEnergyFiltered,'r')
plt.title("kinetic energy to time")
plt.xlabel("time in fs")
plt.ylabel("energy in eV")
plt.show()
```
